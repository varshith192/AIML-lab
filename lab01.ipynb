```python
people=['ann','brandon','chen','david','emily','farook','gagan','hamish','imran','julio']
age=[21,35,45,37,18,52,5,40,48,15]
weight=[55,35,77,68,72,69,18,65,82,48]
height=[160,135,170,165,173,168,175,159,105,171]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight,height)
plt.title("relationship between weight and height of patients")
plt.ylabel("height in meters")
plt.xlabel("weight of patients in kgs")
plt.show()
```

```python
plt.scatter(age,height)
plt.title("relationship between weight and height of patients")
plt.ylabel("age in numbers")
plt.xlabel("height of patients in meters")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.title("people's weight in kgs")
#main plot function 'bar'
plt.bar(x=people,height=weight,width=0.6)
plt.xlabel("x=people",fontsize=15)
#plt.xticks(fontsize=14,rotation=30)
#plt.yticks(fontsize=14)
plt.ylabel("weight (in kgs)")
plt.show
```

```python
plt.figure(figsize=(12,4))
plt.title("people's height in meters")
#main plot function 'bar'
plt.bar(x=people,height=weight,width=0.6)
plt.xlabel("x=people",fontsize=15)
#plt.xticks(fontsize=14,rotation=30)
#plt.yticks(fontsize=14)
plt.ylabel("height (in meters)")
plt.show
```

```python
import numpy as np
plt.figure(figsize=(7,5))
#main plot function 'hist'
plt.hist(weight,color='orange',edgecolor='k',alpha=0.75,bins=5)
plt.title("histogram of patient weight",fontsize=18)
#plt.xticks(fontsize=15)
#plt.yticks(fontsize=15)
plt.show()
```
